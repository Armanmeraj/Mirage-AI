```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/simulated/mass_spring.csv")

plt.figure(figsize=(12, 6))

plt.plot(df["time"], df["position"], label="Position (x)")
plt.plot(df["time"], df["velocity"], label="Velocity (v)")
plt.plot(df["time"], df["acceleration"], label="Acceleration (a)")

plt.title("Mass-Spring System Simulation")
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
omega = 2.0  # same omega used in simulate.py
predicted_acc = -omega**2 * df["position"]
error = (predicted_acc - df["acceleration"]).abs().max()

print(f"Max absolute error between true and expected acceleration: {error:.6f}")
```
