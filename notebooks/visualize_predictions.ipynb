```python
import pandas as pd
import matplotlib.pyplot as plt

# === Load predictions ===
df = pd.read_csv("../results/predictions.csv")

# === Plot true vs predicted acceleration over time ===
plt.figure(figsize=(12, 6))
plt.plot(df["time"], df["acceleration"], label="True Acceleration", linewidth=2)
plt.plot(df["time"], df["predicted_acceleration"], label="Predicted Acceleration", linestyle="--")
plt.xlabel("Time (s)")
plt.ylabel("Acceleration")
plt.title("True vs Predicted Acceleration Over Time")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# === Scatter plot: True vs Predicted ===
plt.figure(figsize=(6, 6))
plt.scatter(df["acceleration"], df["predicted_acceleration"], alpha=0.5)
plt.plot(df["acceleration"], df["acceleration"], color="red", label="Perfect Prediction")
plt.xlabel("True Acceleration")
plt.ylabel("Predicted Acceleration")
plt.title("Predicted vs True Acceleration")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
